```python
import pandas as pd
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
data = pd.read_csv('cleaned_adult_2.csv')

```

```python
# Select the columns for the scatter plot and clustering
x_column = 'Age'
y_column = 'Hours_per_week'
cat_column = 'Work_Class'

# Select the data for clustering
selected_data = data[[x_column, y_column, cat_column]]

# Apply one-hot encoding to the 'Work_Class' column
selected_data = pd.get_dummies(selected_data, columns=[cat_column])

# Apply Gaussian Mixture Model for clustering
n_components = 3  # Specify the number of clusters
gmm = GaussianMixture(n_components=n_components)
gmm.fit(selected_data)

# Get the cluster labels
labels = gmm.predict(selected_data)

# Create the scatter plot with color-coded clusters
plt.figure(figsize=(15, 10))
plt.scatter(data[x_column], data[cat_column], c=labels, cmap='viridis')

# Add labels and title
plt.xlabel(x_column)
plt.ylabel(cat_column)
plt.title('Scatter Plot with Clustering')

# Show the plot
plt.show()
```
