```python
import pandas as pd
import pyfpgrowth

# Load the Adult dataset
adult_df = pd.read_csv('cleaned_adult_2.csv')

# Convert categorical columns to lists of transactions
transactions = adult_df[['Work_Class', 'Education', 'Marital_Status', 
                         'Occupation', 'Relationship', 'Race', 'Sex', 'Native_Country']].values.tolist()

# Define support threshold and confidence threshold
support_threshold = 1  # Adjust as needed
confidence_threshold = 0.5  # Adjust as needed

#The support threshold is the minimum frequency for an itemset to be considered frequent, 
# while the confidence threshold is the minimum confidence for an association rule to be considered strong.

# Perform FP-growth to find frequent itemsets
patterns = pyfpgrowth.find_frequent_patterns(transactions, support_threshold)

# Extract association rules from frequent itemsets
rules = pyfpgrowth.generate_association_rules(patterns, confidence_threshold)

# Print association rules
for rule, confidence in rules.items():
    if len(rule) == 2:  # Ensure the rule has antecedent and consequent
        antecedent = rule[0]
        consequent = rule[1]
        print(f"Rule: {antecedent} -> {consequent}, Confidence: {confidence}")
    else:
        print("Invalid rule:", rule)
```
